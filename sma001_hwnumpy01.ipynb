**Задание 1.** Задана функция `f_python(a, b)`, преобразующая переданные ей два состоящих
из целых чисел списка `a` и `b` одинаковой длины в третий целочисленный список:

~~~
def f_python(a, b):
    return [(a[i] ^ b[i]) * 1_000 + (a[i] | b[i]) // 20 for i in range(len(a))] 
~~~

Объясните подробно (устно), что происходит при следующем вызове этой функции:

~~~
f_python([250], [230])
~~~

Используя библиотеку NumPy и Jupyter Notebook реализуйте функцию
`f_numpy(a, b)`, которая будет выполнять аналогичное преобразование двух
целочисленных массивов, все элементы которых неотрицательны и меньше
50, обеспечивая более быстрое получение результата. 

Напишите далее программный код, позволяющий оценить отношение времён работы 
функций `f_python(a, b)` и `f_numpy(a, b)` на списках/массивах, состоящих
из миллиона полученных с помощью функции
 `random.chоices(range(50), k = 1_000_000)` случайных целых чисел. 

*С целью достижения максимального
ускорения используйте для элементов массива наиболее подходящий тип из

NumPy](https://docs.scipy.org/doc/numpy/user/basics.types.html)
беззнаковых типов для представления целых чисел.*

Для определения времени работы
рекомендуется использовать функцию `time.time()`, а не 
[«магические команды»](https://ipython.readthedocs.io/en/stable/interactive/magics.html)
`%time` или `%timeit`, которые доступны при работе в Jupyter Notebook и с
которыми рекомендуется
[познакомиться](https://ipython.readthedocs.io/en/stable/interactive/magics.html#magic-time).

В результате решения задачи должны быть определены и напечатаны:

- тип элементов массива;
- коэффициент ускорения работы программы после описанной выше модификации,
  округлённый до целого числа.


```python
import numpy as np
from sympy import *
init_printing()
import random
import time
# Определение функции для работы со списками (python)
def f_python(a, b):
    return [(a[i] ^ b[i]) * 1_000 + (a[i] | b[i]) // 20 for i in range(len(a))] 

# Определение функции для работы с массивами (NumPy)
def f_numpy(a, b):
    if all(np.zeros(a.size) <= a) == all(a < np.full(a.size, 50))\
    and all(np.zeros(b.size) <= b) == all(b < np.full(b.size, 50))\
    and a.size == b.size:
        return (a ^ b) * 1_000 + (a | b)//20

# Создание списка 
a = random.choices(range(50), k = 1_000_000)
b = random.choices(range(50), k = 1_000_000)

# Начальный момент времени
t1 = time.time()
f_python(a, b)
# Конечный момент времени
t2 = time.time()
# Разность времен - время выполнения функции
tau_python = t2 - t1

# То же, но для массивов NumPy
c = np.array(a, dtype = np.ushort)
d = np.array(b, dtype = np.ushort)
t1 = time.time()
f_numpy(c, d)
t2 = time.time()
tau_numpy = t2 - t1

# Вывод результатов
print(f"Type of Numpy function's resulting elements of array: {f_numpy(c, d).dtype} \nNumPy function is in {round(tau_python/tau_numpy, 0) } faster then python's one")


```

В результате решения задачи для указанного выше множества данных об
интенсивности потока ы-частиц должны быть вычислены и напечатаны округлённые
до трёх десятичных цифр после запятой:

- среднее значение;
- медиана;
- стандартное отклонение.


```python
# Создаю массив отфильтрованных данных
result = np.array([])

# Фильтрация данных:
# 1. В название файла помещается интересующая нас дата которая преводится из int в str с помощью функции str()
# 2. Из полученного с помощью genfromtxt() массива, выровненного в одну линию с помощью метода ravel()
# берется вырезка соответствующая 15 первым часам, (по условию задачи)
# 3. С помощью функции hsplit() массив разбивается на массив массивов, в каждом из которых по 10 элементов
# соответствующих 10 секундам 
# После этих операций массив принимается за данные, которые нам необходимы
for date in range(22, 28):
    data = np.hsplit(np.genfromtxt("2019-07-" + str(date) + ".csv", delimiter = ',').ravel()[0 : 15 * 3600], 15 * 360)
    
#   print(data[0].size == 10)
#     Массив формируется из реузльтатов усреднения (mean()) каждого массива из 10 элементов в data и стыкуется горизонтально
#     с помощью функции hstack() с самим собой
    result = np.hstack((result, np.array([np.mean(data[i]) for i in range(15 * 360)])))
    
# После работы цикла должны получить массив из 32 400 элементов
# print(result.size == 32400)
# Вывод результатов
print("Среднее значение: ", round(np.mean(result),3))    
print("Медиана: ",round(np.median(result),4))
print("Стандартное отклонение: ",round(np.std(result), 3))
```
